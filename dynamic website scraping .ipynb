```python
from selenium import webdriver
```

```python
from selenium.webdriver.firefox.options import Options as FirefoxOptions
# In Selenium, the Options class is used to configure various options and settings for the Firefox WebDriver.
# It allows you to customize the behavior of the WebDriver, such as enabling headless mode (running the browser without a 
# GUI) or specifying preferences.
from selenium.webdriver.common.by import By
# The By class is used to define different mechanisms for locating elements on a web page. It provides a set of constants
# that you can use as arguments when calling WebDriver's element locating methods (like find_element or find_elements).
# For example, you can use By.ID, By.XPATH, By.CLASS_NAME, etc. to specify how you want to locate elements.
import pandas as pd
```

```python
driver = webdriver.Firefox()

#executable_path = "/Users/kolad/OneDrive/Desktop/web scraping"
# this executable_path is now deprecated, no longer in use
```

```python
url = 'https://www.msport.com/ng/web/sports/list/Soccer?sId=sr:sport:1&d=Today&mId=29'
driver.get(url)
```

```python
#driver.quit()
```

```python
df_list = []
```

```python
tables = driver.find_elements(By.CLASS_NAME, 'm-event')
len(tables)
```

```python
for match_row in tables: # using for loop for iteration
    try:
        time = match_row.find_element(By.CLASS_NAME, 'm-event--header').find_element(By.TAG_NAME, 'span').text
        home_team = match_row.find_element(By.CLASS_NAME, 'm-home-team').text
        away_team = match_row.find_element(By.CLASS_NAME, 'm-away-team').text
        gg_odd = match_row.find_elements(By.CLASS_NAME, 'm-market')[1].find_elements(By.CLASS_NAME, 'odds')[0].text
        ng_odd = match_row.find_elements(By.CLASS_NAME, 'm-market')[1].find_elements(By.CLASS_NAME, 'odds')[1].text
 
        df_dict = {
            'match_time': time,
            'HomeTeam': home_team,
            'AwayTeam': away_team,
            'GG': gg_odd,
            'NG': ng_odd}
    
        df_list.append(df_dict)
        
    except IndexError:
        pass
    


print(df_list)
df = pd.DataFrame(df_list, columns=['match_time', 'HomeTeam', 'AwayTeam', 'GG', 'NG'])
```

```python
msport = df
print(msport)
```
